```python
train['avg_vol_per_month'] = (train['Total Volume Donated (c.c.)'] / train['Number of Donations']) / train['Months since First Donation']
train['avg_wait'] =train['Months since First Donation'] / train['Number of Donations']
train['month_ratio'] = train['Months since First Donation'] / train['Months since Last Donation']
train['month_don'] = train['Number of Donations'] * train['Months since Last Donation']
```
